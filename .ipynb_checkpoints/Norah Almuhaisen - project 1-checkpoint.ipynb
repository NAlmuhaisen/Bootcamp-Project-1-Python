```python

# Function to check for Atherosclerosis
def check_atherosclerosis(symptoms):
    if "Obesity" in symptoms and "Hereditary factor for atherosclerosis" in symptoms and "Blood pressure" in symptoms and "Angina pectoris" in symptoms:
        print("You are suffering from Atherosclerosis.")
        print("Avoid: Obesity, Smoking, Animal fat")
        print("Recommended: Vegetables, fruits, Early treatment, Exercise")

# Function to check for Colon
def check_colon(symptoms):
    if "Lower abdominal pain" in symptoms and "Diarrhoea and dyssynergic defaecation" in symptoms and "Abdominal distention" in symptoms:
        print("You are suffering from Colon.")
        print("Avoid: Sugars and fats, Do not fill the stomach with food")
        print("Be advised: Brown bread, Lactose syrup")

# Function to check for Ulcus
def check_ulcus(symptoms):
    if "Upset stomach" in symptoms and "Nausea" in symptoms and "Severe vomiting" in symptoms and "Blood in the stool" in symptoms and "Drinking alcohol" in symptoms:
        print("You are suffering from Ulcus.")
        print("Avoid: Spice, Fat, Coffee and alcohol, Soft drinks")
        print("Recommended: Milk and low-fat cheese, Hard-boiled eggs, Boiled or grilled meat, Very ripe fruit")

# Main function
def main():
    symptoms = []
    
    print("Please select the symptoms you are experiencing (Enter 'Q' to quit):")
    available_symptoms = [
        "Blood pressure",
        "Angina pectoris",
        "Nausea",
        "Diarrhoea and dyssynergic defaecation",
        "Drinking alcohol",
        "Obesity",
        "Hereditary factor for atherosclerosis",
        "Blood in the stool",
        "Upset stomach",
        "Lower abdominal pain",
        "Severe vomiting",
        "Abdominal distention"
    ]
    
    while True:
        for i, symptom in enumerate(available_symptoms, start=1):
            print(f"{i}. {symptom}")
        
        choice = input("Enter the number of the symptom you are experiencing (or 'Q' to quit): ")
        
        if choice.lower() == 'q':
            break
        
        try:
            choice = int(choice)
            if choice > 0 and choice <= len(available_symptoms):
                symptoms.append(available_symptoms[choice - 1])
        except ValueError:
            print("Invalid input. Please try again.")
    
    if len(symptoms) > 0:
        check_atherosclerosis(symptoms)
        check_colon(symptoms)
        check_ulcus(symptoms)

# Execute the main function
main()



```
