Symptoms Checker Program :

This program allows users to input their symptoms and checks for potential diagnoses such as Atherosclerosis, Colon, or Ulcus based on the selected symptoms.


```python
# Function to check for Atherosclerosis
def check_symptoms(symptoms):
    # To check for Atherosclerosis
    if " Obesity" in symptoms and " Hereditary factor for atherosclerosis" in symptoms and " Blood pressure" in symptoms and " Angina pectoris" in symptoms:
        return("Atherosclerosis")

    # To check for Colon
    elif "Lower abdominal pain" in symptoms and " Diarrhoea and dyssynergic defaecation" in symptoms and "Abdominal distention" in symptoms:
        return("Colon")


    # To check for Ulcus
    elif " Upset stomach" in symptoms and " Nausea" in symptoms and "Severe vomiting" in symptoms and " Blood in the stool" in symptoms and " Drinking alcohol" in symptoms:
        return("Ulcus")


# # Dictionary of symptoms with corresponding numbers
symptoms_dict = {
           '1':" Blood pressure",
        '2':" Angina pectoris",
        '3':" Nausea",
        '4':" Diarrhoea and dyssynergic defaecation",
        '5':" Drinking alcohol",
        '6':" Obesity",
        '7':" Hereditary factor for atherosclerosis",
        '8':" Blood in the stool",
        '9':" Upset stomach",
        '10':"Lower abdominal pain",
        '11':"Severe vomiting",
        '12':"Abdominal distention" 
}

# Explanation of the program
print("Symptoms Checker Program: \n")
print("This program allows users to input their symptoms and checks for potential diagnoses such as Atherosclerosis, Colon, or Ulcus based on the selected symptoms.")

# To print the symptoms list with numbers
print("\nSymptoms List:\n")
for x, y in symptoms_dict.items():
    print(x,".", y)

symptoms = []

break_condition = lambda x: x.lower() == 'q' 

while True:
    choice = input("\nEnter the number of the symptom you are experiencing (one number at a time).\nIf you finish, enter 'Q' to quit : ")

    if break_condition(choice):
            break
            
    # Validate the user input for valid symptom numbers         
    while len(choice) ==0 or int(choice) < 1 or int(choice) > 12 or choice.isdigit()==False : 
        choice = input("\nPlease, enter a valid number of the symptoms list shown above (one number at a time).\nIf you finish, enter 'Q' to quit : ")
    

    if choice in symptoms_dict:
        symptoms.append(symptoms_dict[choice])
  
    # Diagnoses and Recommendations
    if check_symptoms(symptoms)  == "Atherosclerosis":
        print("\nYou are probably suffering from Atherosclerosis.")
        print("Avoid: Obesity, Smoking, Animal fat")
        print("Recommended: Vegetables, fruits, Early treatment, Exercise")
        break

    elif check_symptoms(symptoms)  == "Colon":
        print("\nYou are probably suffering from Colon.")
        print("Avoid: Sugars and fats, Do not fill the stomach with food")
        print("Recommended: Brown bread, Lactose syrup")
        break
        
    elif check_symptoms(symptoms)  == "Ulcus":
        print("\nYou are  probably suffering from Ulcus.")
        print("Avoid: Spice, Fat, Coffee and alcohol, Soft drinks")
        print("Recommended: Milk and low-fat cheese, Hard-boiled eggs, Boiled or grilled meat, Very ripe fruit")
        break

#In case the program couldn't find a matching diagnosis, it will show a message.
if check_symptoms(symptoms) != "Atherosclerosis" and check_symptoms(symptoms) != "Colon" and  check_symptoms(symptoms) != 'Ulcus':
        print ("\nSorry. The system couldn't find a diagnosis matching your symptoms. \nWe suggest seeking medical advice from a doctor.")
```
